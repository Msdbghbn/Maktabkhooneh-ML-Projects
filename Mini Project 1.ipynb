## Data Preprocessing

```python
Data.isnull().sum()
```

```python
Data[Data.Rating.isnull()]
```

```python
Data['Price'] = Data['Price'].str.strip('$')
Data['Price']=Data['Price'].replace(to_replace ='Everyone',value =0)
Data[Data['Price']=='Everyone'].index
Data['Price']=Data['Price'].astype(float)
DataG=Data.groupby(['Content Rating', 'Type'])['Price'].mean()
DataG
```

```python
Data.sort_values( 'Installs',ascending=False)
```

## Strorytelling - Visualization

```python
Data=Data.drop(10472)
Data['Installs'] = Data['Installs'].str.strip('+')
Data['Installs'] = Data['Installs'].str.replace(',','')
Data['Installs']=Data['Installs'].replace(to_replace ='Free',value =0)
Data['Installs']=Data['Installs'].astype(float)
Data
#sns.displot(Data, x="Installs")
```

```python
sns.displot(Data, x="Rating")
```

```python
sns.displot(Data, x="Installs", kind="kde")
```

```python
sns.displot(Data, x="Rating", hue='Type',kind="kde")
```

```python
sns.catplot(x='Content Rating',y='Rating',data=Data)
```

```python
sns.catplot(x="Content Rating", y="Rating", kind="box", data=Data)
```

```python
sns.catplot(x="Content Rating", kind="count", data=Data)
```

```python
sns.relplot(x="Installs", y="Rating", hue='Type',data=Data)
```

```python
Data['Size'] = Data['Size'].str.strip('M')
Data['Size']=Data['Size'].replace(to_replace ='Varies with device',value ='NaN')
Data['Installs']=Data['Installs'].astype(float)
sns.relplot(x="Size", y="Installs", hue='Type',data=Data)
```
