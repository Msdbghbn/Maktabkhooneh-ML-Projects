```python
from pycaret.classification import *
import pandas as pd
```

```python
import io
from google.colab import files
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded['Churn.csv']))
```

```python
classification_setup = setup(data= df, target='Exited',remove_outliers=True,normalize=True,normalize_method='robust',
                            ignore_features= ['Surname', 'CustomerId'], silent = True,session_id = 6563)
```

```python
best=compare_models()
```

```python
top3=compare_models(n_select=3)
```

```python
top3
```

```python
gbc=create_model('gbc')
```

```python
plot_model(estimator=gbc,plot='auc')
```

```python
plot_model(estimator=gbc,plot='feature')
```

```python
rf = create_model('rf');      
lightgbm  = create_model('lightgbm');            

#blending 3 models
blend = blend_models(estimator_list=[gbc,rf,lightgbm])
```

```python
plot_model(estimator=blend,plot='confusion_matrix')
```

```python
pred=predict_model(blend)
```

```python
evaluate_model(blend)
```
